#### 设计规划

界面对象分成两类：信息和任务。根据需求来设计相关的信息和任务，通过组织信息结构、引导用户完成任务设计一系列草图，然后细化草图，得到具体界面，在这个过程中考虑如何让用户完成一个愉悦的故事，这就是设计。为了让用户愉悦，设计任务，我们可能需要当任务复杂时简化操作，当页面信息过多则突出重点。
为了让用户愉悦，设计信息，就是让用户快速的找到自己想要的信息，所以，我们使用导航分类。

分类的两种方式：
+ 根据事物间本身的逻辑关系
+ 根据生活惯例。男装>上装>T恤还是上装>男装>T恤
+ 根据用户想法/tag

导航分类的难点是平衡深度和广度。导航分类，对于程序员来说，本质就是树。通过在界面上显示当前节点的祖父节点，可以让用户明晰自己从哪里来。显示子节点和兄弟节点，让用户知道自己可以去哪里。首页就是根节点，一般首页可以通过暴露两层节点，可以让用户大致了解网站的信息结构。一般深不要超过3层即三次点击，广不要超过7项同时显示

考虑用户习惯的设计：
以淘宝为例，现在“购物车”和“我的淘宝”是以兄弟节点的关系展现在页面上的。但其实，从信息结构上来说“购物车”从属于“我的淘宝”，左侧导航的结构，也体现出这点。但无论对于用户的购买需求，还是网站的促单需求，购物车是至关重要的存在。所以，调整网站的信息架构，为“我的淘宝”增加了兄弟节点“购物车”

比如，我们想在豆瓣上查询最近有什么好看的电影，可以通过导航找到“电影”，再在“电影”频道中查看我们需要的内容并完成相应的任务，如评论、购票。整个过程中，网站需要提供基建，对信息、任务进行分类导航，让用户迅速找到入口，进入到一个具体的界面，再通过该界面完成各种任务。

功能：信息、任务（查看信息、评论、购票）

当有大量的任务/信息需要展示在界面，解决方案：主要任务和次要任务。

需求文档中的功能是扁平化未结构化的，通过确定信息结构把这些功能组织起来，再映射到界面上。

通过信息结构，我们可以把80%的功能组织进来。而通过任务流程，我们则可以把剩下的部分拼装起来。

如果没有任务流程，只是把功能点罗列在界面上，用户就会感到迷惑焦虑。

设计师要理解产品人员的意图，又必须从用户角度思考，确定主任务流，以主任务流为线索，完成产品的整体界面设计。
主任务流确定后，把任务流映射到界面上，形成页面流。页面流就构成了网站的骨架。得到任务页面流设计后，然后再去细化具体的界面

#### 设计标准

原型包含静态的页面样式/线框图和动态的交互说明/操作说明。
原型分为：
+ 低保真：纸面或Axure/Visio软件。
+ 高保真：在外观、操作上与最终产品基本一致，相当于高级模型，更适合做展示。

标准的原型包含：
+ 简要说明和信息结构
+ 任务流程和页面流程
+ 线框图 & 交互说明

简要说明和信息结构：
+ 变更日志表：类似需求文档的变更日志表（字段有：日期、变更内容、变更原因、备注）
+ 版本说明：告诉大家此版本变更了哪些地方（字段有：版本号、日期、更改内容）
+ 信息结构：表明产品范围及其结构方式，信息结构会映射成页面层级和分布。不同于需求文档中的信息结构，需求文档中的信息结构体现的是产品的业务逻辑结构，原型稿中的信息结构，综合考虑了业务逻辑和用户习惯。信息结构的隐含的根节点是功能点//??
